```python
# import libraries
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
# provide the url and read the html
url = 'https://www.sports-reference.com/cbb/boxscores/pbp/2018-11-06-20-michigan.html'
source = requests.get(url).text
soup = BeautifulSoup(source)
```

```python
# get all tbody sections
table = soup.findAll("tbody")

# locate the tbody section we are looking for
i = 0
for entry in table:
    x = str(entry)
    if "data-stat=\"time_remaining\"" in x:
        data = table[i]
    i+= 1

        
pre_df = dict()
desired_columns = ['time_remaining', 'visitor_play', 'visitor_pts', 'score', 
                   'home_pts', 'home_play']
rows = data.find_all('tr')
```

```python
game_log = []

for row in rows:
    df_row = []
    if (row.find('th', {"scope":"row"}) != None):
        for c in desired_columns:
            if c == 'time_remaining':
                cell = row.find("th",{"data-stat": c})
            else:
                cell = row.find("td",{"data-stat": c})
            if cell != None:
                a = cell.text.strip().encode()
                text=a.decode("utf-8")
                df_row.append(text)                
    if (len(df_row) == 6):
        game_log.append(df_row)
```

```python
game_data_cols = ['Time Remaining', 'Visitor Play', 'V_PTS', 'Score', 
                  'H_PTS', 'Home Play']
game_data = pd.DataFrame(game_log, columns = game_data_cols)
game_data
```
